Importing the Dependencies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION & PROCESSING

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# stastical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target variable
data_frame['label'].value_counts()
```

1 ---> Benign
0 ---> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python

X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.2,random_state = 2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logistic Regression model using training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score


```python
 # accuracy on training data
 X_train_prediction = model.predict(X_train)
 training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
 # accuracy on test data
 X_test_prediction = model.predict(X_test)
 test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

Building a Predictive System

```python
import numpy as np

input_data = (18.25,19.98,119.6,1040,0.09463,0.109,0.1127,0.074,0.1794,0.05742,0.4467,0.7732,3.18,53.91,0.004314,0.01382,0.02254,0.01039,0.01369,0.002179,22.88,27.66,153.2,1606,0.1442,0.2576,0.3784,0.1932,0.3063,0.08368
)
# change the input data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The Breast cancer is Malignant')

else:
    print('The Breast cancer is Benign')
```
